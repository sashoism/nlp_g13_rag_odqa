```python
import os
from pathlib import Path
import json
import pandas as pd
```

```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("facebook/contriever")
model = AutoModel.from_pretrained("facebook/contriever")
```

```python
dataset_path = Path("dataset")
```

```python
corpus_path = dataset_path / "wiki_musique_corpus.json"
qa_path = dataset_path / "data"

split = "dev"
test_path = qa_path / f"{split}.json"
```

```python
def tokenize_corpus(corpus_path=corpus_path):
  with open(corpus_path) as fp:
    corpus = json.load(fp)

  passages = [
    "{} SEP {}".format(evidence.get("title"), evidence.get("text"))
    for evidence in corpus.values()
  ]

  tokenized = tokenizer(passages, padding="max_length", max_length=128)

  input_ids = {
    id: input_ids
    for (id, input_ids) in enumerate(tokenized.get("input_ids"))
  }

  attention_mask = {
    id: attention_mask
    for (id, attention_mask) in enumerate(tokenized.get("attention_mask"))
  }

  return {
    "input_ids": input_ids,
    "attention_mask": attention_mask
  }
```

```python
corpus_tokenized_path = "dataset/artifacts/corpus_tokenized.json"
os.makedirs(os.path.dirname(corpus_tokenized_path), exist_ok=True)
```

```python
corpus_tokenized = tokenize_corpus()

with open(corpus_tokenized_path, "w") as fp:
  json.dump(corpus_tokenized, fp)
```
